```python
import pandas as pd
import numpy as np


df=pd.DataFrame({
    "A" : np.append(np.zeros(10), np.ones(10)),
    "B" : np.append(np.zeros(10), np.ones(10))
})
```

```python
df
```

```python
import pandas as pd

row = 0
while row < len(df) - 1:
    current_ = df.iloc[row]
    forward_ = df.iloc[row + 1]
    if (current_ == forward_).all():
        df.drop(index=row + 1, inplace=True)
        df.reset_index(drop=True, inplace=True)  # Reset index after drop
    else:
        row += 1
```
